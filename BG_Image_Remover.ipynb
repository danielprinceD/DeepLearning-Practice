```python
import os
from rembg import remove
from PIL import Image
```

```python
os.makedirs('Original/',exist_ok=True)
os.makedirs('New/',exist_ok=True)
```

```python
img_path = 'Original/image.jpg'
img = Image.open(img_path)
output_path = 'New/removed_img.jpg'
with open(output_path,'wb') as f:
  input = open(img_path,'rb').read()
  subject = remove(input,alpha_matting=True)
  f.write(subject)
```

```python
bg_img = 'Original/bg.jpeg'
new_photo = 'New/merged.jpeg'

bg = Image.open(bg_img)
bg.resize((img.width,img.height))
```

```python
fg = Image.open(output_path)
bg.paste(fg,(0,0),fg)
bg.save(new_photo,format='jpeg')
```
