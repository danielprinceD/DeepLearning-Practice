```python
from PIL import Image
from rembg import remove
import os
os.makedirs('Original',exist_ok=True)
os.makedirs('New',exist_ok=True)
```

```python
img_path = 'Original/anil.jpg'
output_path = 'New/anil.jpg'
with open(output_path,'wb') as f:
  input = open(img_path,'rb').read()
  subject = remove(input,alpha_matting=True)
  f.write(subject)
```

```python
bg_img = 'Original/bg.jpg'
new_photo = 'New/merged.jpg'

img = Image.open(img_path)
bg_img = Image.open(bg_img)
bg_img.resize((img.width,img.height))
```

```python
foreground = Image.open('New/anil.jpg')
bg_img.paste(foreground,(0,0),foreground)
bg_img.save('New/MergedImg.jpg',format='jpeg')
```
